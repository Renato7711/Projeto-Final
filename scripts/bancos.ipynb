```python
#Importando as bibliotecas

import pandas as pd
from sqlalchemy import create_engine
```

```python
engine = create_engine('sqlite:///banco.db', echo=True)
df = pd.read_csv('../basestratadas/arrecadacao-estadov2.csv', sep=',', encoding="utf-8")
df.to_sql('dados', con=engine, if_exists='replace', index=False)
```

```python
df_lido = pd.read_sql('SELECT * FROM dados', con=engine)
df_lido
```

```python
import sqlite3

# Conectar ao banco de dados SQLite
conn = sqlite3.connect('banco.db')

# Ler dados da tabela desejada
query = "SELECT * FROM dados"
df = pd.read_sql_query(query, conn)
df
```
